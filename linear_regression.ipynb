```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
x=df['Y_EXP'].values.reshape(-1, 1)
y=df['SAL'].values.reshape(-1, 1)
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x
Y_std = sc_y

slr = LinearRegression()
slr.fit(X_std, Y_std)
slr.fit(x, y)
        
LinearRegression()
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,slr.predict(x), color='r')
plt.ylabel('salario esperado')
plt.xlabel('promedio de aos')
```

```python
years_exp = 2
years_exp_std=sc_x.transform(np.array([years_exp])).reshape(-1,1)
print("el salario para una dos aos de experiencia  ", sc_y.inverse_transform(slr.predict(years_exp_std)))
```
